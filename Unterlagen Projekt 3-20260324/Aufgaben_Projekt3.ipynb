### a) Tippfeld mit 10 Kästchen 
Die Zufallsvariablen $X_{10}$ beschreibt den Gewinn eines:r Lotterieteilnehmer:in, der/die in einem Tippfeld 10 Kästchen ankreuzt und $2\,€$ einsetzt. Berechnen Sie **Erwartungswert** und **Standardabweichung** von $X_{10}$.

```python
from scipy.stats import hypergeom
import numpy as np
import pandas as pd

# KENO-Parameter
N = 70    # Zahlen gesamt im Tippfeld
K = 20    # davon werden gezogen
einsatz = 2  # Euro

# Gewinntabellen: {Anzahl Richtige: Auszahlung je 1€ Einsatz}
gewinn_tabellen = {
    10: {10: 100000, 9: 1000, 8: 100, 7: 15, 6: 5, 5: 2, 0: 2},
    9:  {9: 50000,  8: 1000, 7: 20,  6: 5,  5: 2, 0: 2},
    8:  {8: 10000,  7: 100,  6: 15,  5: 2,  4: 1, 0: 1},
    7:  {7: 1000,   6: 100,  5: 12,  4: 1},
    6:  {6: 500,    5: 15,   4: 2,   3: 1},
    5:  {5: 100,    4: 7,    3: 2},
    4:  {4: 22,     3: 2,    2: 1},
    3:  {3: 16,     2: 1},
    2:  {2: 6},
}

def berechne_stats(n):
    """Berechnet E, Var, σ für X_n via hypergeometrischer Verteilung."""
    tab = gewinn_tabellen[n]
    k_werte = np.arange(0, n + 1)
    probs   = hypergeom.pmf(k_werte, N, K, n)
    # Netto-Gewinn = Einsatz × Tabellenwert − Einsatz; kein Treffer → -Einsatz
    gewinne = np.array([einsatz * tab[k] - einsatz if k in tab else -einsatz for k in k_werte])
    E   = np.dot(gewinne, probs)
    Var = np.dot((gewinne - E)**2, probs)
    return E, Var, np.sqrt(Var), k_werte, probs, gewinne

# --- a) Vollständige Berechnung für X10 ---
E10, Var10, sig10, k_w, probs, gewinne = berechne_stats(10)

df10 = pd.DataFrame({
    'k (Richtige)':    k_w,
    'P(X10=k)':        probs,
    'Gewinn [€]':      gewinne,
    'Gewinn·P(k)':     gewinne * probs,
    '(G−E)²·P(k)':    (gewinne - E10)**2 * probs,
})
print("Verteilung X10  (10 Tipps, 2 € Einsatz)\n")
print(df10.to_string(index=False, float_format=lambda x: f'{x:.6f}'))
print(f"\nE(X10)   = {E10:.4f} €")
print(f"Var(X10) = {Var10:.4f}")
print(f"σ(X10)   = {sig10:.4f} €")
```

### b) Weitere Tippfelder
Berechnen Sie ebenso die **Erwartungswerte** und **Standardabweichungen** von $X_9, ..., X_2$, d. h. dem Gewinn bei Ankreuzen von 9 (bzw. 8, ..., 2) Kästchen in einem Tippfeld jeweils bei Einsatz von $2\,€$.

***Beispiellösung:***<br>
- Abgabe für a) und b) mit vollständiger Berechnung (nicht nur Endergebnisse)
- am liebsten in tabellarischer Form für $X_{10},...,X_2$

```python
# --- b) Übersicht X9 bis X2 ---
rows = []
for n in range(9, 1, -1):
    E, Var, sig, *_ = berechne_stats(n)
    rows.append({'n': n, 'E(Xn) [€]': round(E, 4), 'Var(Xn)': round(Var, 4), 'σ(Xn) [€]': round(sig, 4)})

df_b = pd.DataFrame(rows).set_index('n')
print("Übersicht X9 bis X2  (jeweils 2 € Einsatz)\n")
print(df_b.to_string())
```

### c) Gewinnchance-Maximierung
Wie viele Kästchen pro Tippfeld sollte ein:e KENO-Spieler:in ankreuzen, der/die den Erwartungswert seines/ihres Gewinns maximieren möchte?

```python
# --- c) Welches n maximiert E(X_n)? ---
e_werte = {n: berechne_stats(n)[0] for n in range(2, 11)}

print("Erwartungswerte im Vergleich:\n")
for n, e in sorted(e_werte.items(), reverse=True):
    marker = " ← Maximum" if n == max(e_werte, key=e_werte.get) else ""
    print(f"  X{n:2d}: E = {e:.4f} €{marker}")

best = max(e_werte, key=e_werte.get)
print(f"\n→ Optimal: {best} Kästchen ankreuzen  (E = {e_werte[best]:.4f} €)")
```

```python
from scipy.stats import binom

# --- d) Endziffer 96 ---
# 12 Tage KENO, täglich 6 Kästchen, 2 € Einsatz
# Verlust wenn k = 0, 1 oder 2 Richtige (kein Eintrag in Gewinntabelle für 6 Tipps)

n_tipps = 6
n_tage  = 12

# Verlustwahrscheinlichkeit an einem Tag (hypergeometrisch)
p_verlust = sum(hypergeom.pmf(k, N, K, n_tipps) for k in [0, 1, 2])
print(f"P(Verlust an einem Tag)       = {p_verlust:.6f}")

# Anzahl Verlusttage ~ Binomial(n=12, p=p_verlust)
# Gesucht: P(genau 9 Verlusttage)
prob = binom.pmf(9, n_tage, p_verlust)
print(f"P(genau 9 Verluste in 12 Tagen) = {prob:.6f}")

print(f"\nFunktion: binom.pmf(k=9, n=12, p={p_verlust:.6f})")
```
